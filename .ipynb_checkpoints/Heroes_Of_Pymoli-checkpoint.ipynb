## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
# Calculate number of unique items
unique_items = purchase_data['Item ID'].value_counts()
ui_df = pd.DataFrame(unique_items)
number_items = ui_df.count()
number_items = number_items.sum() # = 183

# Calculate average price
total_price = pd.Series(purchase_data['Price']).sum()
total_purchases = purchase_data[purchase_data["Price"] != 0]
tp = total_purchases['Price'].count()
average_price = total_price/tp
average_price # = 3.05

# Calculate number of purchases
number_purchases = total_purchases['Price'].count()
number_purchases # = 780

# Calculate total revenue
total_revenue = purchase_data['Price'].sum()
total_revenue # = 2379.77

# Purchase Analysis
purchase_analysis = pd.DataFrame({'Number of Items':[number_items], 'Average Price ($)':[average_price], 
                                  'Number of Purchases':[number_purchases], 'Total Revenue ($)':[total_revenue]})
purchase_analysis.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_bins = [0,9,14,19,24,29,34,39,100]
age_bin_names = ['< 10','10-14','15-19','20-24','25-29','30-34','35-39','40 +']
purchase_data["Age Range"] = pd.cut(purchase_data["Age"], age_bins, labels=age_bin_names)

unique_sn = purchase_data.drop_duplicates(['SN'])
unique_sn.head()

under_ten = unique_sn[unique_sn['Age'] <10]
u10 = under_ten['Age Range'].count()
u10

ten_14 = unique_sn[unique_sn['Age'] < 15]
ten14 = ten_14["Age Range"].count()-u10
ten14

_14_19 = unique_sn[unique_sn['Age'] < 20]
fourteen19 = _14_19["Age Range"].count()-u10-ten14
fourteen19

_20_24 = unique_sn[unique_sn['Age'] < 25]
twenty24 = _20_24["Age Range"].count()-u10-ten14-fourteen19
twenty24

_25_29 = unique_sn[unique_sn['Age'] < 30]
twenty5_29 = _25_29["Age Range"].count()-u10-ten14-fourteen19-twenty24
twenty5_29 

_30_34 = unique_sn[unique_sn['Age'] < 35]
thirty34 = _30_34["Age Range"].count()-u10-ten14-fourteen19-twenty24-twenty5_29
thirty34

_35_39 = unique_sn[unique_sn['Age'] < 40]
thirty5_39 = _35_39["Age Range"].count()-u10-ten14-fourteen19-twenty24-twenty5_29-thirty34
thirty5_39

_40_up = unique_sn[unique_sn['Age'] > 39]
forty_up = _40_up["Age Range"].count()
forty_up

total_count = pd.Series([u10,ten14,fourteen19,twenty24,twenty5_29,thirty34,thirty5_39,forty_up])
total_count

under_ten = (total_count[0]/number_of_players)*100
ten_fourteen = (total_count[1]/number_of_players)*100
fifteen_nineteen = (total_count[2]/number_of_players)*100
twenty_twentyfour = (total_count[3]/number_of_players)*100
twentyfive_twentynine = (total_count[4]/number_of_players)*100
thirty_thirtyfour = (total_count[5]/number_of_players)*100
thirtyfive_thirtynine = (total_count[6]/number_of_players)*100
_forty_up = (total_count[7]/number_of_players)*100

age_range_percent = [under_ten,ten_fourteen,fifteen_nineteen,twenty_twentyfour,twentyfive_twentynine,thirty_thirtyfour,thirtyfive_thirtynine,_forty_up]
age_range_percent

age_demographics = pd.DataFrame({'Total Number':total_count,'Percentage of Players':age_range_percent})
age_demographics
```
